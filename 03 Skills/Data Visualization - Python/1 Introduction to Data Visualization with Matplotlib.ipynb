## Chapter 2.2 Plotting time series with different multiple variables

### Plotting two time-series together

```python
import matplotlib.pyplot as plt
import pandas as pd
# Monthly data from 1958 to 2016
climate_change = pd.read_csv('climate_change.csv',
                             parse_dates=["date"],
                             index_col="date")
climate_change.head()
```

**Index**

climate_change.index

**Plotting the two time-series together**

```python
# Create the plot objects
fig, ax = plt.subplots()

# add the data from var 1 'cos'
ax.plot(climate_change.index, climate_change["co2"])
# add the data from var 2 'relative_temp'
ax.plot(climate_change.index, climate_change["relative_temp"])

# Set labels
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm) / Relative temperature')

# Show the plots
plt.show()
```

- This does not look right, this is because the two plots are on the same scale
- We need to seperate the the scale with `Twin Axis`

### Customizing the Plots

**Using twin axes**

```python
# Create the plot object
fig, ax = plt.subplots()
 
# Add the data 1 'co2' on axis 1
ax.plot(climate_change.index, climate_change["co2"])

# Label the data
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')

# Use "ax.twinx()" to create two seperate axis
ax2 = ax.twinx()

# Add data 2 'relative_temp' onaxis 2
ax2.plot(climate_change.index, climate_change["relative_temp"])
ax2.set_ylabel('Relative temperature (Celsius)')

# Show the plot
plt.show()
```

**Separating variables by color**

```python
# Create a plot object
fig, ax = plt.subplots()

# indicate the var 1 as color='blue'
ax.plot(climate_change.index, climate_change["co2"], color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')

# indicate the var 2 as color='red'
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["relative_temp"], color='red')
ax2.set_ylabel('Relative temperature (Celsius)', color='red')

# Plot
plt.show()
```

**Coloring the ticks - to make them stand out**

```python
# Import Relevant Library
import matplotlib.pyplot as plt

# Create a plot object
fig, ax = plt.subplots()

# Var 1, axis 'ax', color='blue', and tick  colors='blue'
ax.plot(climate_change.index, climate_change["co2"],color='blue')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='blue')
ax.tick_params('y', colors='blue')

# Var 2, new axes 'ax2', color='red' and tick  colors='red'
ax2 = ax.twinx()
ax2.plot(climate_change.index, climate_change["relative_temp"], color='red')
ax2.set_ylabel('Relative temperature (Celsius)',color='red')
ax2.tick_params('y', colors='red')

# Show Plot
plt.show()
```

### Using a function to plots time-series

**Creating the Function**

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)
```

**Using the Function**

```python
# Using our function
fig, ax = plt.subplots()

plot_timeseries(ax, climate_change.index, 
                climate_change['co2'],
                'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()

plot_timeseries(ax, climate_change.index,
                climate_change['relative_temp'],
                'red', 'Time', 'Relative temperature (Celsius)')
plt.show()
```

**Creating re-usable plots with functions**

```python
# Create a function that uses the parameters created from above
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel, color=color)
    axes.tick_params('y', colors=color)
```

```python
# Using our function
# create the plot object
fig, ax = plt.subplots()

plot_timeseries(ax, climate_change.index, climate_change['co2'], 'blue', 'Time', 'CO2 (ppm)')
ax2 = ax.twinx()

plot_timeseries(ax, climate_change.index, climate_change['relative_temp'], 'red', 'Time', 'Relative temperature (Celsius)')
plt.show()
```

## 4.3 Automating figures from data

### How many different kinds of data?

**Our Data**



```python
# Consider this previous data
summer_2016_medals["Sport"]
```

**Getting unique values of a column**

```python
# Unique values of the column
sports = summer_2016_medals["Sport"].unique()
print(sports)
```

**Bar-chart of heights for all sports**

One of the main strengths of Matplotlib is that it can be automated to adapt to the data that it receives as input. For example, if you receive data that has an unknown number of categories, you can still create a bar plot that has bars for each category.

This is what you will do in this exercise. You will be visualizing data about medal winners in the 2016 summer Olympic Games again, but this time you will have a dataset that has some unknown number of branches of sports in it. This will be loaded into memory as a Pandas DataFrame object called summer_2016_medals, which has a column called "Sport" that tells you to which branch of sport each row corresponds. There is also a "Weight" column that tells you the weight of each athlete.

```python
# Barchart to visualise the height of Atheletes in each one of the sports with a standard deviation error bar
fig, ax = plt.subplots()

for sport in sports:
    sport_df = summer_2016_medals[summer_2016_medals["Sport"] == sports]
    ax.bar(sport, sport_df["Height"].mean(),yerr=sport_df["Height"].std())
ax.set_ylabel("Height (cm)")
ax.set_xticklabels(sports, rotation=90)
plt.show()
```

**Figure devived automatically from the graph**



**Automate your visualization**



```python
# Automate your visualization
fig, ax = plt.subplots()

# Loop over the different sports branches
for sport in sports:
  # Extract the rows only for this sport
  sport_df = summer_2016_medals[summer_2016_medals["Sport"] == sport]
  # Add a bar for the "Weight" mean with std y error bar
  ax.bar(sport, sport_df["Weight"].mean(), yerr=sport_df["Weight"].std())

ax.set_ylabel("Weight")
ax.set_xticklabels(sports, rotation=90)

# Save the figure to file
fig.savefig("sports_weights.png")
```
